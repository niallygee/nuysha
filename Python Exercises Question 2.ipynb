```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
plt.rcParams.update({'font.size': 22})
plt.rcParams["figure.figsize"] = (15,10)
```

**a**

```python
t = np.linspace(0, 2, 500)
y1 = np.cos(2*np.pi*10*t + np.pi/6)
plt.plot(t, y1, 'r')
plt.xlabel('Time (s)')
plt.grid(True)
plt.show()
```

***B***

```python
t = np.linspace(0, 2, 500)
y1 = np.cos(2*np.pi*10*t + np.pi/6)
y2 = np.cos(2*np.pi*8*t + np.pi/3)
plt.plot(t, y1, 'r')
plt.plot(t, y2, 'b')
plt.xlabel('Time (s)')
plt.legend(['y1', 'y2'])
plt.grid(True)
plt.show()
```

***C***

```python
t = np.linspace(0, 2, 500)
y3 = y1 + y2
plt.plot(t, y3, 'y')
plt.xlabel('Time (s)')
plt.legend(['y3'])
plt.grid(True)
plt.show()
```

***D***

```python
FFT = np.abs(np.fft.fft(y3)[0:])

FFT2 = FFT[0:250]

plt.plot(FFT2)
plt.grid(True)

plt.subplots(212)
plt.show()
```
